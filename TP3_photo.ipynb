## Comprendre le programme

```python
# ****** Découper l'image *****
from PIL import Image
# Charger le module image depuis PIL
from PIL import ImageFilter
##
img = Image.open('soldat.jpg')
##
img.show()
# Afficher l'image d'origine
print("LxH:", img.size)
# Afficher la définition de l'image d'origine
Xa = 10 ; Ya = 10 ; Xb = 200 ; Yb = 202
##
box = (Xa,Ya,Xb,Yb)
##
decoupe = img.crop(box)
##
decoupe.show()
# Afficher l'image découpée
print("lxh:", decoupe.size)
#
decoupe.save("image_decoupee.png")
# ****** Appliquer un filtre *****
img2 = Image.open("image_decoupee.png")
# Charger l'image découpée
im_filtree = img2.filter(ImageFilter.EDGE_ENHANCE)
# appliquer le filtre
im_filtree.show()
##
im_filtree.save("image_filtree.png")
##
```
